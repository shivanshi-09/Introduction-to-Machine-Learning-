```python
import numpy as np
import pandas as pd
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.kernel_ridge import KernelRidge

dataset = load_diabetes(as_frame=True).frame
dataset.columns
```

```python
x = np.array(dataset.drop('target', axis=1))
y = np.array(dataset['target'])
rng = np.random.default_rng(42)
idx = rng.permutation(len(x))
split = int (0.8*(len(y)))
x_train, x_test = x[idx[:split]] , x[idx[split:]]
y_train, y_test = y[idx[:split]], y[idx[split:]]
```

### Linear Regression
Linear Regression models the relationships between the input and target variable as a linear combination of features and assumes the form:
$$ \hat{y} = \theta^\top x + b $$
where:
- $x$ is the feature vector
- $\theta$ is the vector of the learned coefficients
- $b$ is the intercept term
- $\hat y$ is the predicted output

The parameters $b$ and $\theta$ are estimated by minimizing the mean squared error, which is of the form:

$$ MSE = \frac{1}{n}\sum_{i = 1}^n(y_i- \hat y_i)^2$$


```python
lin_reg = LinearRegression()
lin_reg.fit(x_train, y_train)
y_pred_lin_reg = lin_reg.predict(x_test)
mse_lin_reg = mean_squared_error(y_test, y_pred_lin_reg)
```



### Linear Kernel

The linear kernel corresponds to the standard inner product between two feature vectors:

$$
K(x, x') =  x^T x'
$$

where:
- $x$ and $x'$ are input feature vectors
- $K(x,x')$ represents the similarity between the two vectors.

This kernel does not introduce non-linearity and behaves similarly to standard linear regression.




```python
kr_linear = KernelRidge(kernel="linear", alpha = 0.1)
kr_linear.fit(x_train, y_train)
y_pred_kr_linear = kr_linear.predict(x_test)
mse_kr_linear = mean_squared_error(y_test, y_pred_kr_linear)
```

### Gaussian Kernel

The Gaussian kernel measures similarity between two points based on their Euclidean distance:

$$
K(x, x') = \exp\left(-\gamma \|x - x'\|^2\right)
$$

where:
- $x$ and $x'$ are input feature vectors
- $K(x,x')$ represents the similarity between the two vectors.

The parameter $\gamma$ is defined as

$$
\gamma = \frac{1}{2\sigma^2}
$$

where $\sigma$ controls the width of the Gaussian function. Smaller values of $\sigma$ produce a more localized influence of training samples.

```python
kr_rbf = KernelRidge(kernel="rbf", alpha = 0.1, gamma = 0.1)
kr_rbf.fit(x_train, y_train)
y_pred_kr_rbf = kr_rbf.predict(x_test)
mse_kr_rbf = mean_squared_error(y_test, y_pred_kr_rbf)

print("Linear Regression MSE: ", mse_lin_reg)
print("Linear Kernel MSE: ", mse_kr_linear)
print("Gaussian Kernel MSE: ", mse_kr_rbf)
```
